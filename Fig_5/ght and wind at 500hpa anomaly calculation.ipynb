```python
# @title Heat date list - MAM
list_hw=pd.read_csv('/content/drive/MyDrive/AP_HW/Scripts-ll/HW_dates/data/HW_list_rolling_def.csv')
list_hw['Starting_Date'] = pd.to_datetime(list_hw["Starting_Date"])
list_hw['Ending_Date'] = pd.to_datetime(list_hw["Ending_Date"])

# separate hw for mam season
mam_hw_list =list_hw[(list_hw['Starting_Date'].dt.month >= 3) & (list_hw['Starting_Date'].dt.month <= 5)]
mam_hw_list.reset_index(inplace=True,drop=True)
mam_hw_list
```

```python
# @title Event data for composite analysis

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import matplotlib.colors as colors

#  Initialize empty lists to store results
wind_anom_list = []
ght_anom_list = []

for i in tqdm(range(len(mam_hw_list['Starting_Date']))):
  #  variables
  year=mam_hw_list['Starting_Date'].dt.year.iloc[i]
  start_time =str(mam_hw_list['Starting_Date'][i])
  end_time = str(mam_hw_list['Ending_Date'][i])
  # start_time =np.datetime64(start_time)
  # end_time =np.datetime64(end_time)
  start_time =str(np.datetime64(start_time) - np.timedelta64(6, 'h')) # Substracting 6hrs to convert to BD time.
  end_time =str(np.datetime64(end_time) - np.timedelta64(6, 'h')+ np.timedelta64(24,'h')) # Adding 24hrs to include the end date too.
  time_range=slice(start_time,end_time)
  # print(start_time,end_time)
  event_no=i+1
  ###################################
  ############# geopotential and wind #################
  # Load the files

  ght=xr.open_dataset(f'/content/drive/MyDrive/AP_HW/ERA5 Data/Daily_z_200&500/daily_z_200&500-{year}.nc')
  wind=xr.open_dataset(f'/content/drive/MyDrive/AP_HW/ERA5 Data/Daily_wind_500hpa/daily_uv_500-{year}.nc')
  ght_500_clim=xr.open_dataset('/content/drive/MyDrive/AP_HW/Scripts-ll/climatology/data/z/ght_500_mam_climatology.nc')
  wind_500_clim=xr.open_dataset('/content/drive/MyDrive/AP_HW/Scripts-ll/climatology/data/wind/uv_500_mam_climatology.nc')


  ds_ght=ght.sel(valid_time=time_range,pressure_level=500)
  ds_wind=wind.sel(valid_time=time_range,pressure_level=500)


  ds_ght=ds_ght.mean(dim='valid_time',skipna=True)
  ds_wind=ds_wind.mean(dim='valid_time',skipna=True)

  ght_anom  = ds_ght['z']-ght_500_clim['z']
  wind_anom  = ds_wind-wind_500_clim

  ght_anom=ght_anom.to_dataset()

  ght_anom.coords['event_no'] = event_no
  wind_anom.coords['event_no'] = event_no

  # Append to the list
  ght_anom_list.append(ght_anom)
  wind_anom_list.append(wind_anom)


ght_events_anom_mam = xr.concat(ght_anom_list, dim='event_no')
wind_events_anom_mam = xr.concat(wind_anom_list, dim='event_no')

ght_events_anom_mam.to_netcdf(f'/content/drive/MyDrive/AP_HW/Scripts-ll/ght_wind/data/ght_events_anom-mam.nc')
wind_events_anom_mam.to_netcdf(f'/content/drive/MyDrive/AP_HW/Scripts-ll/ght_wind/data/wind_500_events_anom-mam.nc')
```
